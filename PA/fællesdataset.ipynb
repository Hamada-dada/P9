```python
import pandas as pd
from pathlib import Path

# Unikke stier
BOOKINGER_PATH = Path(r"C:\Users\kfq6\Documents\Data\Bookinger.xlsx")
LABKA_PATH = Path(r"C:\Users\kfq6\Documents\Data\LABKA.xlsx")
POP_PATH = Path(r"C:\Users\kfq6\Documents\Data\Population.xlsx")
WHO5_PATH = Path(r"C:\Users\kfq6\Documents\Data\WHO-5 (PRO).xlsx")

# Læs hver fil én gang, for at undgå gentagen læsning (tage meget hukkeplads)
bookinger_df = pd.read_excel(BOOKINGER_PATH, engine="openpyxl")
labka_df = pd.read_excel(LABKA_PATH, engine="openpyxl")
population_df = pd.read_excel(POP_PATH, engine="openpyxl")
who5_df = pd.read_excel(WHO5_PATH, engine="openpyxl")
```

```python
print(f"Kombineret dataframe: {bookinger_df.shape[0]} rækker, {bookinger_df.shape[1]} kolonner")

print(f"Kombineret dataframe: {labka_df.shape[0]} rækker, {labka_df.shape[1]} kolonner")

print(f"Kombineret dataframe: {population_df.shape[0]} rækker, {population_df.shape[1]} kolonner")

print(f"Kombineret dataframe: {who5_df.shape[0]} rækker, {who5_df.shape[1]} kolonner")

```

```python
# Tilføj kolonne for kilde
bookinger_df["Kilde"] = "Bookinger"
labka_df["Kilde"] = "LABKA"
population_df["Kilde"] = "Population"
who5_df["Kilde"] = "WHO-5 (PRO)"

# Konkatter alle sammen
combined_df = pd.concat(
    [bookinger_df, labka_df, population_df, who5_df],
    ignore_index=True
)

print(f"Kombineret dataframe: {combined_df.shape[0]} rækker, {combined_df.shape[1]} kolonner")
print(combined_df["Kilde"].value_counts())
```
