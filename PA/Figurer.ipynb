```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
from pathlib import Path

RAW_DATA = """Region,Periode,Persons,Population,IndicatorReported,PrevalencePct,MedianAge
DK,Q1 2015,29625,5668743,29625,0.52,46.2
Nordjylland,Q1 2015,3029,582664,3029,0.52,44.4
DK,Q2 2015,29728,5678348,29728,0.52,46.2
Nordjylland,Q2 2015,3044,583471,3044,0.52,44.4
DK,Q3 2015,29802,5699220,29802,0.52,46.2
Nordjylland,Q3 2015,3048,584918,3048,0.52,44.4
DK,Q4 2015,29928,5707251,29928,0.52,46.1
Nordjylland,Q4 2015,3069,585499,3069,0.52,44.3
DK,Q1 2016,29982,5717014,29982,0.52,46.1
Nordjylland,Q1 2016,3075,585769,3075,0.52,44.3
DK,Q2 2016,30040,5724456,30040,0.52,46.1
Nordjylland,Q2 2016,3087,586104,3087,0.53,44.3
DK,Q3 2016,30113,5745526,30113,0.52,46.1
Nordjylland,Q3 2016,3095,587421,3095,0.53,44.4
DK,Q4 2016,30175,5748769,30175,0.52,46.0
Nordjylland,Q4 2016,3094,587335,3094,0.53,44.4
DK,Q1 2017,30220,5756170,30220,0.53,45.9
Nordjylland,Q1 2017,3093,587446,3093,0.53,44.4
DK,Q2 2017,30380,5760694,30380,0.53,45.9
Nordjylland,Q2 2017,3130,587611,3130,0.53,44.4
DK,Q3 2017,30490,5778570,30490,0.53,46.0
Nordjylland,Q3 2017,3154,588962,3154,0.54,44.3
DK,Q4 2017,30590,5781190,30590,0.53,45.9
Nordjylland,Q4 2017,3167,589148,3167,0.54,44.4
DK,Q1 2018,30706,5785864,30706,0.53,45.9
Nordjylland,Q1 2018,3197,588766,3197,0.54,44.3
DK,Q2 2018,30817,5789957,30817,0.53,45.9
Nordjylland,Q2 2018,3216,588888,3216,0.55,44.4
DK,Q3 2018,30878,5806015,30878,0.53,45.9
Nordjylland,Q3 2018,3215,590010,3215,0.54,44.5
DK,Q4 2018,31023,5806081,31023,0.53,45.9
Nordjylland,Q4 2018,3221,589755,3221,0.55,44.5
DK,Q1 2019,31316,5811413,31316,0.54,45.9
Nordjylland,Q1 2019,3266,589605,3266,0.55,44.6
DK,Q2 2019,31692,5814461,31692,0.55,46.1
Nordjylland,Q2 2019,3329,589731,3329,0.56,44.9
DK,Q3 2019,31924,5827463,31924,0.55,46.2
Nordjylland,Q3 2019,3380,590580,3380,0.57,45.2
DK,Q4 2019,32276,5822763,32276,0.55,46.3
Nordjylland,Q4 2019,3431,589936,3431,0.58,45.5
DK,Q1 2020,32540,5824857,32540,0.56,46.3
Nordjylland,Q1 2020,3441,589712,3441,0.58,45.7
DK,Q2 2020,32710,5825337,32710,0.56,46.4
Nordjylland,Q2 2020,3482,589837,3482,0.59,45.8
DK,Q3 2020,32857,5837213,32857,0.56,46.4
Nordjylland,Q3 2020,3511,590739,3511,0.59,46.0
DK,Q4 2020,33057,5840045,33057,0.57,46.4
Nordjylland,Q4 2020,3538,590439,3538,0.60,46.1
DK,Q1 2021,33260,5843347,33260,0.57,46.4
Nordjylland,Q1 2021,3556,590322,3556,0.60,46.2
DK,Q2 2021,33451,5850189,33451,0.57,46.5
Nordjylland,Q2 2021,3577,590403,3577,0.61,46.2
DK,Q3 2021,33615,5867412,33615,0.57,46.5
Nordjylland,Q3 2021,3600,591724,3600,0.61,46.2
DK,Q4 2021,33795,5873420,33795,0.58,46.5
Nordjylland,Q4 2021,3613,591740,3613,0.61,46.3
DK,Q1 2022,33968,5883562,33968,0.58,46.5
Nordjylland,Q1 2022,3648,591758,3648,0.62,46.4
DK,Q2 2022,34150,5910577,34150,0.58,46.5
Nordjylland,Q2 2022,3648,594323,3648,0.61,46.5
DK,Q3 2022,34337,5928364,34337,0.58,46.6
Nordjylland,Q3 2022,3699,594925,3699,0.62,46.5
DK,Q4 2022,34502,5932654,34502,0.58,46.5
Nordjylland,Q4 2022,3707,594634,3707,0.62,46.4
DK,Q1 2023,34703,5941388,34703,0.58,46.5
Nordjylland,Q1 2023,3726,594426,3726,0.63,46.4
DK,Q2 2023,34855,5944145,34855,0.59,46.5
Nordjylland,Q2 2023,3742,594016,3742,0.63,46.4
DK,Q3 2023,34944,5959464,34944,0.59,46.5
Nordjylland,Q3 2023,3759,593898,3759,0.63,46.5
DK,Q4 2023,35039,5961249,35039,0.59,46.6
Nordjylland,Q4 2023,3761,593135,3761,0.63,46.5
DK,Q1 2024,35197,5967824,35197,0.59,46.6
Nordjylland,Q1 2024,3768,592831,3768,0.64,46.5
DK,Q2 2024,35309,5972420,35309,0.59,46.6
Nordjylland,Q2 2024,3776,592637,3776,0.64,46.6
DK,Q3 2024,35413,5989985,35413,0.59,46.7
Nordjylland,Q3 2024,3787,593260,3787,0.64,46.6
DK,Q4 2024,35575,5992734,35575,0.59,46.7
Nordjylland,Q4 2024,3819,592768,3819,0.64,46.6
DK,Q1 2025,35678,5999677,35678,0.59,46.7
Nordjylland,Q1 2025,3836,592394,3836,0.65,46.6
"""

def quarter_to_timestamp(qstr: str) -> pd.Timestamp:
    """Convert 'Q# YYYY' to end-of-quarter Timestamp."""
    q, year = qstr.split()
    qnum = int(q[1:])
    year = int(year)
    return pd.Period(f"{year}Q{qnum}").to_timestamp(freq='Q')

def main(output_dir: str = "."):
    # Load data
    df = pd.read_csv(StringIO(RAW_DATA))
    df["Date"] = df["Periode"].apply(quarter_to_timestamp)
    df = df.sort_values(["Date", "Region"]).reset_index(drop=True)

    # Persons per 100,000 population
    df["Per100k"] = (df["Persons"] / df["Population"]) * 100_000

    # Pivot for plotting
    rate = df.pivot(index="Date", columns="Region", values="Per100k")

    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(rate.index, rate["DK"], label="Denmark (national)")
    plt.plot(rate.index, rate["Nordjylland"], label="Region Nordjylland")
    plt.title(T"Type 1 Diabetes per 100,000 Population, Denmark vs. Region Nordjylland (2015â€“2025)")
    plt.xlabel("Quarter")
    plt.ylabel("Persons per 100,000")
    plt.legend()
    plt.tight_layout()

    # Save outputs
    out_dir = Path(output_dir)
    out_dir.mkdir(parents=True, exist_ok=True)
    pdf_path = out_dir / "t1dm_per100k_DK_vs_Nordjylland_2015_2025.pdf"
    png_path = out_dir / "t1dm_per100k_DK_vs_Nordjylland_2015_2025.png"
    plt.savefig(pdf_path)          # vector (best for LaTeX)
    plt.savefig(png_path, dpi=200) # raster
    plt.close()

    print(f"Saved:\n  {pdf_path}\n  {png_path}")

if __name__ == "__main__":
    # Set a custom output directory here if you like, e.g. "1.Problemanalysis/Figurer"
    main(output_dir=".")
```
